```python
from PIL import Image, ImageDraw

width, height = 500, 500
image = Image.new('RGB',(width,height),(0,0,0))
draw = ImageDraw.Draw(image)

max_it = 100 
xmin, xmax = -2, 2        #can be used to 'zoom' into function
dx = xmax - xmin
ymin, ymax = -1, 1
dy = ymax - ymin

def Julia(z):
    z = z
    c = complex(-0.97, 0.2)      #set from assignment
    i = 0 
    while i < max_it and abs(z) <= 2:    #abs(z) set by looking for best picture results
        z = z**2 +c
        i += 1
    return i


for x in range(width):
    for y in range(height):
        z = complex(x / width * dx + xmin,y / height * dy + ymin)
        m = Julia(z)
        color = 255 - int(m*255/max_it)
        draw.point([x,y],(color,color,color))
image.show()
```

```python
width, height = 500, 500
image = Image.new('RGB',(width,height),(0,0,0))
draw = ImageDraw.Draw(image)

max_it = 100
amin, amax = -2, 2
da = amax - amin
bmin, bmax = -1, 1
db = bmax - bmin

def Mand(c):
    z = 0
    i = 0 
    while i < max_it and abs(z) <= 10:
        z = z**4 + c
        i += 1
    return i


for a in range(width):
    for b in range(height):
        c = complex(a / width * da + amin,b / height * db + bmin)
        m = Mand(c)
        color = 255 - int(m*255/max_it)
        draw.point([a,b],(color,color,color))
image.show()
```
